```python
from comet import multiverse
from functions.multiverse_integration import compare_methods, plot_methods, integrate
from pathlib import Path
# 1. Load multiverse results and plot
DATA_DIR = Path.cwd()  # Current directory (examples/hurricane_multiverse_non-neuroimaging)
mverse = multiverse.load_multiverse(str(DATA_DIR))
df = mverse.get_results(as_df=True, expand_dec=True)
mverse.multiverse_plot(measure="beta", n_bins=20, sig_col="p_value", baseline=0, figsize=(7,9))
#print(df.columns.tolist())- to see available measures

# 2. Compare methods and get weights
table, weights = compare_methods(df, measure="beta")
print(table)

# 3. Plot weighted densities
fig, ax = plot_methods(df, measure="beta", weights=weights)
```
